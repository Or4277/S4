```python
l = ['a','b','a','b','a','b','b']
```

```python
nombre_a = 0
nombre_b = 0
for lettre in l:
    if lettre == 'a':
        nombre_a +=1
    else :
        nombre_b +=1
if nombre_b == nombre_a:
    print('meme nombre')
else: 
    print('pas le meme nombre')
```
